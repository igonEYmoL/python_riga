```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/mpg_ggplot2.csv')

df.head()
```

```python
mean_hwy = df['hwy'].mean()
print("Mean highway miles per gallon:", mean_hwy)
max_speed = df['hwy'].max()
print("Maximum highway miles per gallon:", max_speed)
min_speed = df['hwy'].min()
print("Minimum highway miles per gallon:", min_speed)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['hwy'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Highway Miles per Gallon')
plt.xlabel('Highway MPG')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10, 6))
plt.scatter(df['displ'], df['hwy'], alpha=0.5, color='blue')
plt.title('Engine Displacement vs. Highway MPG')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Highway MPG')
plt.grid(True)
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
